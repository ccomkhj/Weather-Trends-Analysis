```python
from utils import load_weather, average, compare
```

### 1. Between two departments, set up the correct batch start and end
Batch end means the date you want to see the data

```python
start_date = "2023-02-01"
end_date = "2023-04-01"
new_start_date = "2024-02-01"
new_end_date = "2024-04-01"
latitude = 49.4521
longitude = 11.0767
```

### 2. Load weather using geographic info

```python
weather = load_weather(start_date, end_date, latitude, longitude)
new_weather = load_weather(new_start_date, new_end_date, latitude, longitude)
```

### 3. Process data: keep only moving average

```python
avg_weather = average(weather, weeks=4)
new_avg_weather = average(new_weather, weeks=4)
```

### 4. Visualization

```python
compare(avg_weather, new_avg_weather)
```
